```python
#Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_with_discounts.csv')
```

```python
numerical_columns = df.select_dtypes(include=['number']).columns
```

```python
print(numerical_columns)
```

```python
means = df[numerical_columns].mean()
```

```python
modes = df[numerical_columns].mode().iloc[0]
```

```python
medians = df[numerical_columns].median()
```

```python
std_devs = df[numerical_columns].std()
```

```python
stats = pd.DataFrame({'Mean': means, 'Median': medians, 'Mode': modes, 'Standard Deviation': std_devs})
print(stats)
```

```python
#Data Visualisation
```

```python
import matplotlib.pyplot as plt
```

```python
df[numerical_columns].hist(figsize=(10, 10), bins=20)
plt.show()
```

```python
df[numerical_columns].plot(kind='box', subplots=True, figsize=(10, 10))
plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_columns)
for col in categorical_columns:
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

```python
#Standardization of Numerical Variables
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
print(df[numerical_columns].head())
```

```python
#Conversion of Categorical Data into Dummy Variables
```

```python
df = pd.get_dummies(df, columns=categorical_columns)
print(df.head())
```

```python
#Summarize key findings and reflect on the importance of data preprocessing:
```

Descriptive Analytics: Provided insights into central tendency and dispersion of numerical columns.

Data Visualization: Helped understand distributions and identify outliers.

Standardization: Ensured numerical columns are on a similar scale, which is crucial for machine learning models.

One-hot Encoding: Transformed categorical data into a numerical format suitable for ML algorithms.
